```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python

df = pd.read_csv("/content/Housing.csv")
print(df.head())
```

```python

df = df.dropna()


X_simple = df[['area']]
y = df['price']


X_multiple = df[['bedrooms', 'bathrooms', 'area']]
y = df['price']
```

```python
# Simple regression
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_simple, y, test_size=0.2, random_state=42)

# Multiple regression
X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_multiple, y, test_size=0.2, random_state=42)
```

```python
# Simple regression
lr_simple = LinearRegression()
lr_simple.fit(X_train_s, y_train_s)

# Multiple regression
lr_multiple = LinearRegression()
lr_multiple.fit(X_train_m, y_train_m)
```

```python
# For simple regression
y_pred_s = lr_simple.predict(X_test_s)

print("Simple Linear Regression Metrics:")
print("MAE:", mean_absolute_error(y_test_s, y_pred_s))
print("MSE:", mean_squared_error(y_test_s, y_pred_s))
print("RMSE:", np.sqrt(mean_squared_error(y_test_s, y_pred_s)))
print("R² Score:", r2_score(y_test_s, y_pred_s))

# For multiple regression
y_pred_m = lr_multiple.predict(X_test_m)

print("\nMultiple Linear Regression Metrics:")
print("MAE:", mean_absolute_error(y_test_m, y_pred_m))
print("MSE:", mean_squared_error(y_test_m, y_pred_m))
print("RMSE:", np.sqrt(mean_squared_error(y_test_m, y_pred_m)))
print("R² Score:", r2_score(y_test_m, y_pred_m))
```

```python
# Plotting regression line (simple)
plt.scatter(X_test_s, y_test_s, color='blue', label='Actual')
plt.plot(X_test_s, y_pred_s, color='red', linewidth=2, label='Predicted')
plt.xlabel('Sqft Living')
plt.ylabel('Price')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

```python
print("Simple Regression Coefficient:", lr_simple.coef_)
print("Simple Regression Intercept:", lr_simple.intercept_)

print("\nMultiple Regression Coefficients:", lr_multiple.coef_)
print("Multiple Regression Intercept:", lr_multiple.intercept_)
```
